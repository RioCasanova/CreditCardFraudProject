```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('../../datasets/credit-card-fraud-dataset/fraudTrain.csv')
```

```python
# printing the first five rows in the dataset
credit_card_data.head()
```

```python
# Printing the last 5 rows of the dataset
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()
```

```python
# checking for missing values in the dataset in each column
credit_card_data.isnull()
```

```python
# checking the distribution of legit transactions & fraud transactions
# it is good to know the balance of the dataset because if the dataset is
# unbalanced it may not be a good set to teach with
credit_card_data['is_fraud'].value_counts()
```

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.is_fraud == 0]
fraud = credit_card_data[credit_card_data.is_fraud == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.amt.describe()
```

```python
# statistical measures of the fraudulent data
fraud.amt.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('is_fraud').mean()
```

```python
# Build a sample dataset containing similar distribution of normal 
# transactions and fraudulent transactions 
legit_sample = legit.sample(n=7506)
```

```python
# Concatenating two dataframes
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
# we are going to find the value count of fraud and normal trans
new_dataset['is_fraud'].value_counts()
```

```python
# This is a good way to check if we got a good sample or a bad sample
# We can see that the means still correspond to the differences with 
# the larger samples.
new_dataset.groupby('is_fraud').mean()
```

```python
# Splitting the data into features and targets
# Here we are creating a new dataset that does not have the 'is_fraud'
# column and we are creating a variable that has only the 'is_fraud' info
X = new_dataset.drop(columns='is_fraud',axis=1)
Y = new_dataset['is_fraud']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Split the data into Training Data and Testing Data
# This will be split randomly
# 'test_size': is the percentage of the data that you want to use - 
#     80% goes to the training data and 20% to the testing data
# 'X_train': is the dataset that does not contain the 'is_fraud' data
# 'Y_train': is what contains all 'is_fraud' data as labels
# 'stratify=Y': evenly distributes the Y data over the training and test data
# 'Random_state': How you want to split your data

# The way I understand it is: There are four variables, 80% of the data
#  goes into the training data which is the variables ending in 'train'
#  and 20% goes into test. The variables that start with 'Y' is the data
#  that corresponds to the 'is_fraud' data which tells which transaction
#  is or isnt fraud. Variables starting with 'X' incldues the rest of 
#  the data. This is essentially grouping by the 'is_fraud' column.
#  This all makes sense when you realize it is supposed to go on a graph.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model Training -------------------------
#  Logistic Regression
```

```python
model = LogisticRegression()
```

```python
# Training the LogisticRegression Model with training data
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
#  Accuracy Scrore
#   accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```
